```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate a 100-sample long synthetic test signal from the model

```python
f = 0.017
sample = 100

x = np.arange(sample)
w = np.sqrt(0.25) * np.random.randn(sample)
y = np.sin(2 * np.pi * f * x) + w
```

```python
plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Estimate sinusoidal parameters

```python
scores = []
frequencies = []
n = np.arange(100)
for f in np.linspace(0, 0.5, 1000):
    z = -2 * np.pi * 1j * f * n
    e = np.exp(z)
    score = np.abs(np.dot(y, e))
    scores.append(score)
    frequencies.append(f)
frequencies[np.argmax(scores)]
```
